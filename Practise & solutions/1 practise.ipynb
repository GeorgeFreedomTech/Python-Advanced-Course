```python
# practise I

# 1.
class Operative:
    def __init__(self, callsign: str, clearance_level: int, specialization: str, contact_channel: str):
        self.callsign = callsign
        self.clearance_level = clearance_level
        self.specialization = specialization
        self.contact_channel = contact_channel

# 2. 
operative_alpha = Operative("Helios", 5, "Astro-cartography", "SEC-COM-1")
operative_beta = Operative("Raptor", 4, "Deep Range Recon", "TAC-COM-3")

print(operative_alpha.specialization)
print(operative_beta.specialization)


# 3. 
priority_target_list = ["Helios", "Vesper", "Orion"]
list_of_operatives = [operative_alpha, operative_beta]

for operative in list_of_operatives:
    if operative.callsign in priority_target_list:
        print(f"Alert: Priority target '{operative.callsign}' detected.")
    else:
        print(f"Operative '{operative.callsign}' is not on the priority list.")


# practise II 

class LabAssistant:
    def __init__(self, name: str, id_code: str, location: str):
        self.name = name
        self.id = id_code
        self.location = location
        self.known_protocols = [] # Initialized as an empty list

    def report_status(self):
        print(f"Assistant '{self.name}' (ID: {self.id}) is currently at: {self.location}.")

    def relocate(self, new_location: str):
        self.location = new_location

    def learn_protocol(self, new_protocol: str):
        self.known_protocols.append(new_protocol)

# 3. 
    def compare_skills(self, other_assistant):
        print(f"Robot {self.name} knows {len(self.known_protocols)} protocols.")
        print(f"Robot {other_assistant.name} knows {len(other_assistant.known_protocols)} protocols.")

# 4. 
    def compare_skills_extended(self, other_assistant):
        self_count = len(self.known_protocols)
        other_count = len(other_assistant.known_protocols)

        print(f"Robot {self.name} knows {self_count} protocols.")
        print(f"Robot {other_assistant.name} knows {other_count} protocols.")
        
        if self_count > other_count:
            print(f"Robot {self.name} is more experienced than Robot {other_assistant.name}.")
        elif self_count < other_count:
            print(f"Robot {other_assistant.name} is more experienced than Robot {self.name}.")
        else:
            print("Both robots have the same experience level.")


# --- Example Usage ---

bot_rx1 = LabAssistant("RX-1", "LAB-AI-001", "Main Laboratory")
bot_z5 = LabAssistant("Z-5", "LAB-AI-002", "Sample Analysis Wing")


bot_rx1.learn_protocol("Sample Handling")
bot_rx1.learn_protocol("Data Encryption")

bot_z5.learn_protocol("Quantum Comms")
bot_z5.learn_protocol("High-Energy Containment")
bot_z5.learn_protocol("Robotic Arm Calibration")

bot_rx1.relocate("Storage Bay C")
bot_rx1.report_status()

print("\n--- Skill Comparison ---")
bot_z5.compare_skills(bot_rx1)
```
